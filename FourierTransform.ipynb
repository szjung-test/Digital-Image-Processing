## python 푸리에변환 np.fft.fft() 함수 이용

```python
import numpy as np
import matplotlib.pyplot as plt

fs = 1000
t = np.arange(0, 5, 1 /fs)
# 40Hz 를 갖는 0.3 진폭 신호1, 8Hz를 갖는 3 진폭 신호2 생성 후 두 신호 합하여 신호 생성
f1 = 40
f2 = 8
signal = 0.3 * np.sin(2 * np.pi * f1 * t) + 3 * np.cos(2 * np.pi * f2 * t + np.pi/2)

# 생성된 신호를 푸리에 변환 ->  함수의 반환값(fft) 신호의 길이로 나눔 = normalization 
# fft 결과는 복소수
fft = np.fft.fft(signal) / len(signal)

# normalization을 절대값 취함 
fft_magnitude = abs(fft)
```

```python
print(fft_magnitude)
```

생성된 신호의 푸리에 변환 결과 값

```python
plt.subplot(2,1,1)
plt.plot(t, signal)
plt.grid()

plt.subplot(2,1,2)
plt.stem(fft_magnitude)
plt.ylim(0,2.5)
plt.grid()
plt.show()
```

- O(x) : Odd function(기함수, anti-symmetrical, odd)
